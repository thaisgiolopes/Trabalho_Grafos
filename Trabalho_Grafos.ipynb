### EstatÃ­sticas dos problemas

Para visualizar as estatÃ­sticas calculadas dos problemas, execute o cÃ³digo abaixo:

```python
estatisticas_gerais = []
intermediacoes = []

# LÃª os nomes dos arquivos
with open("lista_instancias.txt", "r") as arquivo:
    nomes_instancias = [linha.strip() for linha in arquivo if linha.strip()]

# Processa cada instÃ¢ncia
for nome_arquivo in nomes_instancias:
    caminho = f"instancias/{nome_arquivo}"
    grafo = Grafo(caminho)
    grafo.gerar_matriz_caminhos()

    estatisticas, intermed = grafo.calcular_estatisticas()
    estatisticas["instancia"] = nome_arquivo
    estatisticas_gerais.append(estatisticas)

    for no, valor in intermed.items():
        intermediacoes.append({
            "instancia": nome_arquivo,
            "vertice": no,
            "centralidade_intermediacao": valor
        })

# Cria DataFrames
df_estatisticas = pd.DataFrame(estatisticas_gerais)
# Reordena as colunas para que "instancia" fique em primeiro
colunas = ["instancia"] + [col for col in df_estatisticas.columns if col != "instancia"]
df_estatisticas = df_estatisticas[colunas]
df_intermediacao = pd.DataFrame(intermediacoes)

# Exibe tabelas
print("ðŸ“Š EstatÃ­sticas Gerais por InstÃ¢ncia:")
display(df_estatisticas)

print("ðŸ“Š Centralidade de IntermediaÃ§Ã£o por VÃ©rtice:")
display(df_intermediacao)

# Salva como CSV
df_estatisticas.to_csv("estatisticas_gerais.csv", index=False)
df_intermediacao.to_csv("centralidade_intermediacao.csv", index=False)
```
